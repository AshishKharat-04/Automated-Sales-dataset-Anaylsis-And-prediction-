# Regional Sales Analysis - Data Cleaning & Transformation



## 1. Import Libraries

```python
# Data manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Utilities
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.float_format', '{:.2f}'.format)

print("Libraries imported successfully!")
```

## 2. Load Dataset

```python
# Load all sheets from the Excel file
excel_file = 'Regional_Sales_Dataset_2021_2024.xlsx'
sheets = pd.read_excel(excel_file, sheet_name=None)

# Assign dataframes to each sheet
df_sales = sheets['Sales Orders']
df_customers = sheets['Customers']
df_products = sheets['Products']
df_regions = sheets['Regions']
df_state_reg = sheets['State Regions']
df_budgets = sheets['2022 Budgets']

print("Dataset loaded successfully!")
print(f"\nNumber of sheets: {len(sheets)}")
print(f"Sheet names: {list(sheets.keys())}")
```

## 3. Initial Data Exploration

```python
# Function to display dataset info
def explore_dataframe(df, name):
    print(f"\n{'='*80}")
    print(f"Dataset: {name}")
    print(f"{'='*80}")
    print(f"Shape: {df.shape}")
    print(f"\nColumns: {list(df.columns)}")
    print(f"\nData Types:\n{df.dtypes}")
    print(f"\nMissing Values:\n{df.isnull().sum()}")
    print(f"\nFirst 3 rows:")
    display(df.head(3))

# Explore each dataset
explore_dataframe(df_sales, "Sales Orders")
explore_dataframe(df_customers, "Customers")
explore_dataframe(df_products, "Products")
explore_dataframe(df_regions, "Regions")
explore_dataframe(df_state_reg, "State Regions")
explore_dataframe(df_budgets, "2022 Budgets")
```

## 4. Data Cleaning - Sales Orders

```python
# Create a copy for cleaning
df_sales_clean = df_sales.copy()

print("Original Sales Data:")
print(f"Shape: {df_sales_clean.shape}")
print(f"\nMissing values:\n{df_sales_clean.isnull().sum()}")
```

```python
# 4.1 Handle Date Columns
df_sales_clean['OrderDate'] = pd.to_datetime(df_sales_clean['OrderDate'], errors='coerce')

# Check date range
print(f"Date Range: {df_sales_clean['OrderDate'].min()} to {df_sales_clean['OrderDate'].max()}")
print(f"Missing dates: {df_sales_clean['OrderDate'].isna().sum()}")

# Remove rows with missing dates
df_sales_clean = df_sales_clean.dropna(subset=['OrderDate'])
print(f"\nRows after removing missing dates: {len(df_sales_clean)}")
```

```python
# 4.2 Extract Date Features
df_sales_clean['Year'] = df_sales_clean['OrderDate'].dt.year
df_sales_clean['Month'] = df_sales_clean['OrderDate'].dt.month
df_sales_clean['Quarter'] = df_sales_clean['OrderDate'].dt.quarter
df_sales_clean['Day'] = df_sales_clean['OrderDate'].dt.day
df_sales_clean['DayOfWeek'] = df_sales_clean['OrderDate'].dt.dayofweek
df_sales_clean['DayName'] = df_sales_clean['OrderDate'].dt.day_name()
df_sales_clean['MonthName'] = df_sales_clean['OrderDate'].dt.month_name()
df_sales_clean['WeekOfYear'] = df_sales_clean['OrderDate'].dt.isocalendar().week

print("Date features extracted successfully!")
print(f"\nNew columns: {['Year', 'Month', 'Quarter', 'Day', 'DayOfWeek', 'DayName', 'MonthName', 'WeekOfYear']}")
```

```python
# 4.3 Handle Missing Values in Numeric Columns
numeric_cols = ['Order Quantity', 'Unit Price', 'Line Total', 'Total Unit Cost']

for col in numeric_cols:
    missing_count = df_sales_clean[col].isna().sum()
    if missing_count > 0:
        print(f"{col}: {missing_count} missing values")
        # Fill with median
        df_sales_clean[col].fillna(df_sales_clean[col].median(), inplace=True)

print("\nMissing numeric values handled!")
```

```python
# 4.4 Remove Duplicates
print(f"Duplicate rows: {df_sales_clean.duplicated().sum()}")
df_sales_clean = df_sales_clean.drop_duplicates()
print(f"Rows after removing duplicates: {len(df_sales_clean)}")
```

```python
# 4.5 Calculate Profit Margin
df_sales_clean['Profit'] = df_sales_clean['Line Total'] - df_sales_clean['Total Unit Cost']
df_sales_clean['Profit Margin %'] = (df_sales_clean['Profit'] / df_sales_clean['Line Total']) * 100

print("Profit metrics calculated!")
print(f"\nProfit Summary:")
print(df_sales_clean[['Line Total', 'Total Unit Cost', 'Profit', 'Profit Margin %']].describe())
```

```python
# 4.6 Handle Outliers (Optional - using IQR method)
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Check outliers in Line Total
outliers, lower, upper = detect_outliers_iqr(df_sales_clean, 'Line Total')
print(f"Outliers in Line Total: {len(outliers)} ({len(outliers)/len(df_sales_clean)*100:.2f}%)")
print(f"Bounds: [{lower:.2f}, {upper:.2f}]")

# Keep outliers but flag them
df_sales_clean['Is_Outlier'] = ((df_sales_clean['Line Total'] < lower) | 
                                 (df_sales_clean['Line Total'] > upper)).astype(int)
print(f"\nOutliers flagged in 'Is_Outlier' column")
```

## 5. Data Cleaning - Supporting Tables

```python
# 5.1 Clean Customers Data
df_customers_clean = df_customers.copy()
df_customers_clean = df_customers_clean.drop_duplicates()
df_customers_clean['Customer Names'] = df_customers_clean['Customer Names'].str.strip()

print(f"Customers cleaned: {len(df_customers_clean)} unique customers")
```

```python
# 5.2: Clean Products Data
df_products_clean = df_products.copy()
df_products_clean = df_products_clean.drop_duplicates()

# Clean product names (column is 'Product Name', not 'Product Description')
if 'Product Name' in df_products_clean.columns:
    df_products_clean['Product Name'] = df_products_clean['Product Name'].str.strip()

print(f"Products cleaned: {len(df_products_clean)} unique products")
print(f"Columns: {df_products_clean.columns.tolist()}")
```

```python
# 5.3 Clean Regions Data
df_regions_clean = df_regions.copy()
df_regions_clean = df_regions_clean.drop_duplicates()

# Handle missing values in regions
print(f"Missing values in Regions:\n{df_regions_clean.isnull().sum()}")

# Fill missing numeric values with 0 or median
numeric_region_cols = df_regions_clean.select_dtypes(include=[np.number]).columns
for col in numeric_region_cols:
    df_regions_clean[col].fillna(df_regions_clean[col].median(), inplace=True)

print(f"\nRegions cleaned: {len(df_regions_clean)} unique regions")
```

## 6. Merge Data for Analysis

```python
# 6.1 Merge Sales with Customers
df_merged = df_sales_clean.merge(
    df_customers_clean,
    left_on='Customer Name Index',
    right_on='Customer Index',
    how='left'
)

print(f"After merging with customers: {df_merged.shape}")
```

```python
# Section 6.2: Merge with Products
# Note: Products sheet has 'Index' and 'Product Name' columns (not 'Product Index')
df_merged = df_merged.merge(
    df_products_clean,
    left_on='Product Description Index',
    right_on='Index',  # ← CORRECTED: was 'Product Index', now 'Index'
    how='left',
    suffixes=('', '_product')
)

# Rename for clarity
if 'Product Name' in df_merged.columns:
    df_merged.rename(columns={'Product Name': 'Product Description'}, inplace=True)

print(f"After merging with products: {df_merged.shape}")
```

```python
# 6.3 Merge with Regions
df_merged = df_merged.merge(
    df_regions_clean,
    left_on='Delivery Region Index',
    right_on='id',
    how='left',
    suffixes=('', '_region')
)

print(f"After merging with regions: {df_merged.shape}")
print(f"\nFinal merged columns: {len(df_merged.columns)}")
```

## 7. Final Data Quality Checks

```python
# Check final data quality
print("Final Merged Dataset Summary:")
print(f"{'='*80}")
print(f"Total Rows: {len(df_merged):,}")
print(f"Total Columns: {len(df_merged.columns)}")
print(f"\nDate Range: {df_merged['OrderDate'].min()} to {df_merged['OrderDate'].max()}")
print(f"\nYears covered: {sorted(df_merged['Year'].unique())}")
print(f"\nMissing Values:")
missing = df_merged.isnull().sum()
print(missing[missing > 0])

print(f"\nData Types:")
print(df_merged.dtypes.value_counts())
```

## 8. Create Analysis-Ready Datasets

```python
# Section 8.1: Sales Analysis Dataset (Main)
# Define desired columns - using actual column names from the data
sales_analysis_cols = [
    'OrderNumber', 'OrderDate', 'Year', 'Month', 'Quarter', 'MonthName',
    'Customer Names', 'Channel', 'Product Description', 
    'Warehouse Code', 'state', 'name', 'Order Quantity', 'Unit Price',
    'Line Total', 'Total Unit Cost', 'Profit', 'Profit Margin %',
    'Is_Outlier'
]

# Select columns that exist
available_cols = [col for col in sales_analysis_cols if col in df_merged.columns]
df_sales_analysis = df_merged[available_cols].copy()

print(f"Sales Analysis Dataset created with {len(df_sales_analysis)} rows and {len(available_cols)} columns")
print(f"\nColumns included: {available_cols}")
```

```python
# 8.2 Monthly Aggregated Dataset (for time series)
df_monthly = df_merged.groupby(['Year', 'Month']).agg({
    'Line Total': 'sum',
    'Profit': 'sum',
    'Order Quantity': 'sum',
    'OrderNumber': 'count'
}).reset_index()

df_monthly.columns = ['Year', 'Month', 'Revenue', 'Profit', 'Quantity', 'Orders']
df_monthly['Year-Month'] = pd.to_datetime(df_monthly[['Year', 'Month']].assign(Day=1))
df_monthly = df_monthly.sort_values('Year-Month')

print(f"Monthly dataset created with {len(df_monthly)} months of data")
print(df_monthly.head())
```

```python
# 8.3 Product Performance Dataset
if 'Product Description' in df_merged.columns:
    df_product_performance = df_merged.groupby('Product Description').agg({
        'Line Total': 'sum',
        'Profit': 'sum',
        'Order Quantity': 'sum',
        'OrderNumber': 'count'
    }).reset_index()
    
    df_product_performance.columns = ['Product', 'Revenue', 'Profit', 'Quantity', 'Orders']
    df_product_performance['Avg_Order_Value'] = df_product_performance['Revenue'] / df_product_performance['Orders']
    df_product_performance = df_product_performance.sort_values('Revenue', ascending=False)
    
    print(f"Product performance dataset created with {len(df_product_performance)} products")
    print(df_product_performance.head())
```

```python
# 8.4 Regional Performance Dataset
if 'state' in df_merged.columns:
    df_regional_performance = df_merged.groupby('state').agg({
        'Line Total': 'sum',
        'Profit': 'sum',
        'Order Quantity': 'sum',
        'OrderNumber': 'count'
    }).reset_index()
    
    df_regional_performance.columns = ['State', 'Revenue', 'Profit', 'Quantity', 'Orders']
    df_regional_performance = df_regional_performance.sort_values('Revenue', ascending=False)
    
    print(f"Regional performance dataset created with {len(df_regional_performance)} states")
    print(df_regional_performance.head())
```

## 9. Export Cleaned Data

```python
# 9. Export Cleaned Data to CSV and XLSX
print("\n" + "="*60)
print("Exporting datasets...")

# Export to CSV
df_sales_analysis.to_csv('cleaned_sales_data.csv', index=False)
df_monthly.to_csv('monthly_aggregated_data.csv', index=False)

if 'df_product_performance' in locals():
    df_product_performance.to_csv('product_performance.csv', index=False)

if 'df_regional_performance' in locals():
    df_regional_performance.to_csv('regional_performance.csv', index=False)

print("✓ CSV files exported!")

# Export to Excel (single file with multiple sheets)
with pd.ExcelWriter('Cleaned_Sales_Data.xlsx', engine='openpyxl') as writer:
    df_sales_analysis.to_excel(writer, sheet_name='Sales Data', index=False)
    df_monthly.to_excel(writer, sheet_name='Monthly Aggregated', index=False)
    
    if 'df_product_performance' in locals():
        df_product_performance.to_excel(writer, sheet_name='Product Performance', index=False)
    
    if 'df_regional_performance' in locals():
        df_regional_performance.to_excel(writer, sheet_name='Regional Performance', index=False)

print("✓ Excel file exported: Cleaned_Sales_Data.xlsx")

print("\nExported files:")
print("CSV Files:")
print("  1. cleaned_sales_data.csv")
print("  2. monthly_aggregated_data.csv")
print("  3. product_performance.csv")
print("  4. regional_performance.csv")
print("\nExcel File (all sheets in one file):")
print("  • Cleaned_Sales_Data.xlsx")
print("="*60)
```
